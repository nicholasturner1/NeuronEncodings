```python
soma_syn_ids = []
for cell_id in soma_ids:

    if put_type[cell_id] != "pyramidal":
        continue

    cell_syn_ids = np.where(syns[:, 1] == cell_id)[0]
    cell_syns = syns[cell_syn_ids]
        
    try:
        skel_nodes, skel_edges, _, skel_root = su.load_postprocessed_skeleton(cell_id, node_dist_nm, v2=v2, blanc=True)
    except:
        print("No skeleton found for %d" % cell_id)
        continue
        
    compartment_pred = np.load(path_dict["save_folder_ax_maj"] + "soma_%d_el_%d_fc_4000_smiter_0.npy" % (cell_id, node_dist_nm))

    skel_kdtree = spatial.cKDTree(skel_nodes) 
    _, syn_to_skel_id = skel_kdtree.query(cell_syns[:, 2:5], k=10)
        
    soma_syn_ids.extend(cell_syn_ids[np.any(compartment_pred[syn_to_skel_id] == 2, axis=1)])

soma_syn_ids = np.array(soma_syn_ids, dtype=np.int)
```

```python
syns_on_soma = syns[soma_syn_ids]
syns_not_on_soma = syns[~np.in1d(np.arange(len(syns), dtype=np.int), soma_syn_ids)]
syns_not_on_soma = syns_not_on_soma[np.in1d(syns_not_on_soma[:, 1], soma_ids)]
```

```python
n_soma_syns = []
n_syns = []
avg_contact_areas = []
std_contact_areas = []
n_contacts = []
avg_syn_sizes = []
std_syn_sizes = []
avg_syn_size_soma = []
std_syn_size_soma = []
avg_syn_size_non_soma = []
std_syn_size_non_soma = []
avg_syn_n = []
std_syn_n = []
avg_syn_n_per_soma = []
std_syn_n_per_soma = []
avg_syn_n_per_non_soma = []
std_syn_n_per_non_soma = []
n_mitos = []
avg_mito_size = []
std_mito_size = []
mito_per_syn_size = []

time_start = time.time()
for i_bc, bc in enumerate(u_basket_cells):
    if i_bc % 20 == 1:
        dt = time.time() - time_start
        eta = dt / i_bc * len(u_basket_cells) - dt
        print("%4d / %4d - dt: %.3fs - eta: %.3fs" % (i_bc, len(u_basket_cells), dt, eta), end="\r")
    
    bc_syn_ids = np.where(syns[:, 0] == bc)[0]
    bc_syns = syns[bc_syn_ids]
    bc_soma_syns = bc_syns[np.in1d(bc_syn_ids, soma_syn_ids)]
    bc_non_soma_syns = bc_syns[~np.in1d(bc_syn_ids, soma_syn_ids)]
    
    n_soma_syns.append(len(bc_soma_syns))
    n_syns.append(len(bc_syns))
    
    avg_syn_sizes.append(np.mean(bc_syns[:, -1]))
    std_syn_sizes.append(np.std(bc_syns[:, -1]))
    
    avg_syn_n.append(np.mean(np.unique(bc_syns[:, 1], return_counts=True)[1]))
    std_syn_n.append(np.std(np.unique(bc_syns[:, 1], return_counts=True)[1]))
    avg_syn_n_per_soma.append(np.mean(np.unique(bc_soma_syns[:, 1], return_counts=True)[1]))
    std_syn_n_per_soma.append(np.std(np.unique(bc_soma_syns[:, 1], return_counts=True)[1]))
    
    avg_syn_size_soma.append(np.mean(bc_soma_syns[:, -1]))
    std_syn_size_soma.append(np.std(bc_soma_syns[:, -1]))
    
    if len(bc_non_soma_syns) == 0:
        avg_syn_n_per_non_soma.append(-1)
        std_syn_n_per_non_soma.append(-1)
        avg_syn_size_non_soma.append(-1)
        std_syn_size_non_soma.append(-1)
    else:
        avg_syn_n_per_non_soma.append(np.mean(np.unique(bc_non_soma_syns[:, 1], return_counts=True)[1]))
        std_syn_n_per_non_soma.append(np.std(np.unique(bc_non_soma_syns[:, 1], return_counts=True)[1]))
        avg_syn_size_non_soma.append(np.mean(bc_non_soma_syns[:, -1]))
        std_syn_size_non_soma.append(np.std(bc_non_soma_syns[:, -1]))
    
    contact_areas = contacts[:, -1][np.logical_or(contacts[:, 0] == bc, contacts[:, 1] == bc)]
    n_contacts.append(len(contact_areas))
    
    avg_contact_areas.append(np.mean(contact_areas))
    std_contact_areas.append(np.std(contact_areas))
    
    if bc in seg_to_mitos_dict:
        bc_mitos = mitos[np.in1d(mitos[:, 0], seg_to_mitos_dict[bc]), -1]
        avg_mito_size.append(np.mean(bc_mitos))
        std_mito_size.append(np.mean(bc_mitos))
        mito_per_syn_size.append(np.sum(bc_mitos) / len(bc_syns))
        n_mitos.append(len(bc_mitos))
    else:
        avg_mito_size.append(-1)
        std_mito_size.append(-1)
        mito_per_syn_size.append(-1)
        n_mitos.append(0)
        
    
n_syns = np.array(n_syns)
n_soma_syns = np.array(n_soma_syns)
avg_contact_areas = np.array(avg_contact_areas)
std_contact_areas = np.array(std_contact_areas)
avg_syn_sizes = np.array(avg_syn_sizes)
std_syn_sizes = np.array(std_syn_sizes)
avg_syn_size_soma = np.array(avg_syn_size_soma)
std_syn_size_soma = np.array(std_syn_size_soma)
avg_syn_size_non_soma = np.array(avg_syn_size_non_soma)
std_syn_size_non_soma = np.array(std_syn_size_non_soma)
n_contacts = np.array(n_contacts)
avg_syn_n = np.array(avg_syn_n)
std_syn_n = np.array(std_syn_n)
avg_syn_n_per_soma = np.array(avg_syn_n_per_soma)
std_syn_n_per_soma = np.array(std_syn_n_per_soma)
avg_syn_n_per_non_soma = np.array(avg_syn_n_per_non_soma)
std_syn_n_per_non_soma = np.array(std_syn_n_per_non_soma)
n_mitos = np.array(n_mitos)
avg_mito_size = np.array(avg_mito_size)
std_mito_size = np.array(avg_mito_size)
mito_per_syn_size = np.array(mito_per_syn_size)
```

```python
soma_syn_ratios = n_soma_syns / n_syns
syn_contact_ratios = n_syns / n_contacts
syn_mito_ratios = n_syns / n_mitos
syn_mito_ratio_mean = np.mean(syn_mito_ratios[~np.logical_or(np.isnan(syn_mito_ratios), np.isinf(syn_mito_ratios))])
syn_mito_ratios[np.isnan(syn_mito_ratios)] = syn_mito_ratio_mean
syn_mito_ratios[np.isinf(syn_mito_ratios)] = syn_mito_ratio_mean
```

```python
def replace_with_average(arr):
    arr[arr == -1] = np.mean(arr[arr != -1])
    return arr
```

```python
avg_syn_n_per_non_soma = replace_with_average(avg_syn_n_per_non_soma)
std_syn_n_per_non_soma = replace_with_average(std_syn_n_per_non_soma)
avg_syn_size_non_soma = replace_with_average(avg_syn_size_non_soma)
std_syn_size_non_soma = replace_with_average(std_syn_size_non_soma)

avg_mito_size = replace_with_average(avg_mito_size)
std_mito_size = replace_with_average(std_mito_size)
mito_per_syn_size = replace_with_average(mito_per_syn_size)
```
