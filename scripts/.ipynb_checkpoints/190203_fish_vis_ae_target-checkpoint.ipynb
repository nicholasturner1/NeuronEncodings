## Latent encoding

```python
cell_views = []
cell_view_seg_ids = []
for seg_id in seg_ids: 
    print(seg_id)
    mesh = mesh_meta.mesh(seg_id=seg_id, merge_large_components=False)
    
    if mesh.n_vertices < n_pts:
        continue
        
    views, centers = ne.pred.encode_mesh_by_views(model, mesh,
                                                  n_points=n_pts,
                                                  sample_n_points=n_pts,
                                                  fisheye=False,
                                                  batch_size=batch_size,
                                                  order="random",
                                                  pc_align=True,
                                                  method="kdtree",
                                                  max_samples=200)
    cell_views.append(views)
    cell_view_seg_ids.append(seg_id)
    
views_flat = np.concatenate([c for c in cell_views if len(c) > 0])
cell_view_seg_ids = np.array(cell_view_seg_ids)
```

```python
features_t, explained_variance_ratio_, pca = unsupervised.get_pca_components(views_flat, normed=True)
bins = [np.histogram(features_t[:, i], bins=10)[1] for i in range(len(features_t[0]))]

cell_view_summary = []

loc = 0
for cell_view in cell_views:
    if len(cell_view) == 0:
        continue
        
    cell_view_t = pca.transform(cell_view)

    hist_feats = []
    
#     for i_dim in range(len(cell_view_t[0])):
#         hist_feats.extend(np.histogram(cell_view_t[:, i_dim], bins=bins[i_dim])[0])
    
    hist_feats = np.array(hist_feats)
    
    cell_view_summary.append(np.concatenate([np.min(cell_view_t, axis=0), 
                                             np.max(cell_view_t, axis=0), 
                                             np.std(cell_view_t, axis=0), 
                                             np.mean(cell_view_t, axis=0),
                                             np.median(cell_view_t, axis=0),
                                             hist_feats]))
    loc += len(cell_view)
    
cell_view_summary = np.array(cell_view_summary)
```

```python
features_t, explained_variance_ratio_, pca = unsupervised.get_pca_components(cell_view_summary, normed=True)

cell_view_summary_n = np.array(cell_view_summary).copy()
# std = np.std(cell_view_summary_n, axis=0)[None]
# std[std == 0] = 1
# cell_view_summary_n /= std
# cell_view_summary_n -= np.nanmean(cell_view_summary_n, axis=0)[None]
kmeans = cluster.KMeans(n_clusters=10)
classes = kmeans.fit_predict(cell_view_summary_n)

for dim in range(0, 9):
    fig = plt.figure(figsize=(8, 8))
    
    for c in np.unique(classes):
        lw = 0
        if c > 9:
            lw = 1
            
        plt.scatter(features_t[classes == c, dim], features_t[classes == c, dim+1], alpha=.7, label=c, lw=lw, edgecolor="k")
        
    plt.legend()
    plt.show()
```

```python
for c in np.unique(classes):
    ids = np.where(classes == c)[0]
    np.random.shuffle(ids)
    print(c, cell_view_seg_ids[ids][:200])
#     print(compute_cluster_difference(cell_view_seg_ids[ids]))
```

```python
def compute_cluster_difference(seg_ids):
    locs = []
    for seg_id in seg_ids: 
        mesh = mesh_meta.mesh(seg_id=seg_id, merge_large_components=False)
        locs.append(mesh.vertices[np.argmin(mesh.vertices[:, 1])])
    
    pdists = spatial.distance.pdist(locs)
    return np.mean(pdists)
```
